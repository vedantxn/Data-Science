Project: Interactive Data Visualization with Fancy Indexing
ðŸ”¹ Goal: Use Fancy Indexing to highlight and manipulate specific data points in a scatter plot.

ðŸ”¹ Concepts Used: NumPy Fancy Indexing, Boolean Masking, Matplotlib for Visualization.

ðŸ”¹ Tools Required: Python, NumPy, Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 10  # X values (0 to 10)
y = np.random.rand(100) * 10  # Y values (0 to 10)

# Identify the top 10 highest 'y' values using Fancy Indexing
top_10_indices = np.argsort(y)[-10:]  # Get indices of top 10 y-values

# Apply Boolean Masking to filter points where x > 5
mask = x > 5

# Plot all data points
plt.scatter(x, y, color='blue', alpha=0.5, label="All Points")

# Highlight the top 10 y-values
plt.scatter(x[top_10_indices], y[top_10_indices], color='red', label="Top 10 y-values", edgecolors="black", s=100)

# Highlight points where x > 5
plt.scatter(x[mask], y[mask], color='green', label="x > 5", edgecolors="black", s=50)

# Customize the plot
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Data Visualization with Fancy Indexing")
plt.legend()
plt.grid(True)
plt.show()
```
